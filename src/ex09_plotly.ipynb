## Exercise 09 : Plotly

Do what you need to do to create a graph like this:



```python
import pandas as pd
import sqlite3
import plotly.graph_objects as go
import numpy as np
con= sqlite3.connect("data/checking-logs.sqlite")
main_query="""
SELECT
    uid,
    DATE(timestamp) AS date,
    numTrials
FROM checker 
WHERE uid LIKE "user_%" AND status=='ready' AND labname=='project1'
ORDER BY date, uid
"""
df=pd.read_sql(main_query, con, parse_dates=['date'])
df
```

```python

users = df['uid'].unique()
dates = df['date'].dt.date.unique()

frames = []
for current_date in dates:
    frame_data = df[df['date'].dt.date <= current_date]
    data_for_frame = []
    for user in users:
        user_data = frame_data[frame_data['uid'] == user]
        trace = go.Scatter(
            x=user_data['date'],
            y=user_data['numTrials'],
            mode='lines',
            name=user
        )
        data_for_frame.append(trace)
    frames.append(go.Frame(
        data=data_for_frame,
        name=str(current_date)
    ))

initial_data = []
for user in users:
    initial_data.append(go.Scatter(
        x=[],
        y=[],
        mode='lines',
        name=user
    ))

figure = go.Figure(
    data=initial_data,
    layout=go.Layout(
        title="Dynamic of commits per user in project1",
        xaxis=dict(title="Date", range=[df['date'].min(), df['date'].max()]),
        yaxis=dict(title="numTrials", range=[0, df['numTrials'].max()]),
        showlegend=True,
        legend=dict(
            title='uid',
            x=1.02,
            y=1,
            bgcolor='rgba(255, 255, 255, 0.5)',
            bordercolor='Black',
            borderwidth=1
        ),
        updatemenus=[dict(
            type="buttons",
            buttons=[
                dict(
                    label="Play",
                    method="animate",
                    args=[None, {"frame": {"duration": 100, "redraw": False}, "fromcurrent": True}]
                )
            ]
        )],
        sliders=[dict(
            steps=[
                dict(
                    method='animate',
                    args=[[str(date)], {"frame": {"duration": 100, "redraw": False}, "mode": "immediate"}],
                    label=str(date)
                ) for date in dates
            ],
            transition={"duration": 30},
            x=0,
            xanchor="left",
            y=0,
            yanchor="top"
        )]
    ),
    frames=frames
)
figure.update_layout(template='ggplot2')
figure
```
